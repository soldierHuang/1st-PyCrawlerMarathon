```python
# 1. é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ

import requests
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response = r.text

import json
data = json.loads(response)

item = 0
column = []

#å›å‚³å¹¾ç­†è³‡æ–™
for d in data:
    item +=1
    # print(i,end=' / ')
print("item count = ",item) 

#æ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½
for d in data[0]:
  column.append(d)
print("column list = ",column)
```

```python
# 2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€

import pandas as pd

popular = []
item = 0
count = 0
counts = 0
for d in data:
  item += 1
  for j in d["reactions"]:
    count += int(j["count"])
    # counts += count
    # print(item ,count,counts)
  popular.append( [ d['title'],d['createdAt'],d['commentCount'],d['likeCount'],count ] )

df = pd.DataFrame(popular,columns=['æ¨™é¡Œ','è²¼æ–‡æ™‚é–“','ç•™è¨€äººæ•¸','æŒ‰è®šäººæ•¸','reactions'])
df
```
