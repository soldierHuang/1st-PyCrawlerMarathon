### 讀讀看「test」資料夾下的圖檔！

```python
# !pip install pytesseract
# !sudo apt install tesseract-ocr

import pytesseract
```

```python
# 先下載圖片，再存取，解析成文字 "cv2_imshow"

# !curl -o test1.png http://pycrawler.cupoy.com/file-download/part32/test1.png

from google.colab.patches import cv2_imshow
import cv2

test1_png = cv2.imread('test1.png')
code = pytesseract.image_to_string(test1_png)
print(f'show code :',code)
cv2_imshow(test1_png)
```

```python
# 讀取一張圖片，解析成文字 "PIL Image"
from PIL import Image
from io import BytesIO
import requests

response = requests.get('http://pycrawler.cupoy.com/file-download/part32/test1.png')
test1_get = Image.open(BytesIO(response.content))
code = pytesseract.image_to_string(test1_get)
print(f'show code :',code)
```
