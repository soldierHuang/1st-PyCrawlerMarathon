## 1. 取出今天所有的新聞

```python
# !pip install selenium
# !apt install chromium-chromedriver

from selenium import webdriver

options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')

browser = webdriver.Chrome('chromedriver',options=options)
browser.maximize_window()

from bs4 import BeautifulSoup
import requests
import time

browser.get("https://www.ettoday.net/news/news-list.htm")

for i in range(500):
    browser.execute_script('window.scrollTo(0,100000)')
    time.sleep(0.01)
```

```python
import pandas as pd
import datetime

df = []

today = datetime.date.today().strftime('%Y/%m/%d')

html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
      date = d.find_all('span')[-1].text.split(' ')[0]
      time = d.find_all('span')[-1].text.split(' ')[1]
      category = d.find_all('em')[-1].text
      title = d.find_all('a')[-1].text

      if date == today:
          # print(date,time,category,title)
          df.append([date,time,category,title])

df = pd.DataFrame(df,columns=['date','time','category','title'])
df
```

## 2. 取出現在時間兩小時內的新聞

```python
x = datetime.datetime.now().strftime('%H')

adjust = 2   # hours adjust range from now(system time)

# result = []

for i in range(adjust):
    x_i = str(int(x)-i).zfill(2)
    print(df[df.time.str.contains(f'({x_i}):.')])
    # result.append(df[df.time.str.contains(f'({x_i}):.')])

# result
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
from selenium.webdriver.support.ui import Select

# browser.get("https://www.ettoday.net/news/news-list.htm")

today = datetime.datetime.now().timetuple()
year = str(today[0])
month = str(today[1])
day = str(today[2]-3)  #3天前
day_before = year+"/"+month+"/"+ day


selectYear = Select(browser.find_element_by_id("selY"))
selectMonth = Select(browser.find_element_by_id("selM"))
selectDay = Select(browser.find_element_by_id("selD"))
search_btn = browser.find_element_by_id('button')

selectMonth.select_by_value(month)
selectYear.select_by_value(year)
selectDay.select_by_value(day)
search_btn.click()


import time

for i in range(500):
    browser.execute_script('window.scrollTo(0,100000)')
    time.sleep(0.01)


html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")

df1 = []

for d in soup.find(class_="part_list_2").find_all('h3'):
      date = d.find_all('span')[-1].text.split(' ')[0]
      time = d.find_all('span')[-1].text.split(' ')[1]
      category = d.find_all('em')[-1].text
      title = d.find_all('a')[-1].text
      # print(date,time,category,title)

      if date == day_before:
        df1.append([date,time,category,title])

df1 = pd.DataFrame(df1,columns=['date','time','category','title'])

df1[df1.time.str.contains('(15|16):.+|17:00')]
```
