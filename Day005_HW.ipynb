## 作業目標

* 請利用 API: https://www.dcard.tw/_api/forums/pet/posts?popular=true 回答下列問題：

1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？

import requests
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response = r.text

import json
data = json.loads(response)

item = 0
column = []

#回傳幾筆資料
for d in data:
    item +=1
    # print(i,end=' / ')
print("item count = ",item) 

#每一筆資料包含哪些欄位
for d in data[0]:
  column.append(d)
print("column list = ",column)
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」

import pandas as pd

popular = []
item = 0
count = 0
counts = 0
for d in data:
  item += 1
  for j in d["reactions"]:
    count += int(j["count"])
    # counts += count
    # print(item ,count,counts)
  popular.append( [ d['title'],d['createdAt'],d['commentCount'],d['likeCount'],count ] )

df = pd.DataFrame(popular,columns=['標題','貼文時間','留言人數','按讚人數','reactions'])
df
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

hot_rank = df.sort_values(by='reactions',ascending=False)
# hot_rank
print("熱門文章 平均")
print(hot_rank.head(5).mean())
print()
print("非熱門文章 平均")
hot_rank.tail(5).mean()
```
