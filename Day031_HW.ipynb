### 尋找一個網站是有作 Header 檢查的（Hint: 有加跟沒有加回傳結果不一樣）

```python
import requests

url = 'https://www.cwb.gov.tw/' 
r = requests.get(url)
response = r.text

response
```

```python
import requests

url = 'https://www.cwb.gov.tw/V8/C/' 

headers = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.117 Safari/537.36',
    'authority': 'www.cwb.gov.tw',
    'method': 'GET',
    'accept': '*/*',
    'path': '/V8/C/ajax/_home_news.html?_=1578645510521',
    'scheme': 'https',
    'accept-language': 'zh-TW,zh;q=0.9,en-US;q=0.8,en;q=0.7',
    'cookie': 'V8_LANG=C',
    'x-requested-with': 'XMLHttpRequest',
    'sec-fetch-mode': 'cors',
    'sec-fetch-site': 'same-origin',
    'referer': 'https://www.cwb.gov.tw/V8/C/'
        }


r = requests.get(url,headers=headers)
r.encoding='utf-8'
response = r.text

response
```

### 用上述網站說明該如何判斷 Header 中必須加上哪些資料才會正確


```python
import requests

url = 'https://www.cwb.gov.tw/V8/C/' 

headers = ''


r = requests.get(url,headers=headers)
r.encoding='utf-8'
response = r.text


response
```
