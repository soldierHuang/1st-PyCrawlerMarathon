## 作業目標

根據範例 ，完成以下問題：

* ① 印出最新文章的「作者」「標題」「時間」
* ② 印出第一頁所有文章的「作者」「標題」「時間」


### ① 印出最新文章的「作者」「標題」「時間」

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ptt.cc/bbs/NBA/index.html'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")

for d in soup.find_all(class_="title"):
    print(d.text.replace('\t', '').replace('\n', ''))
    try:
        r = BeautifulSoup(requests.get('https://www.ptt.cc'+d.find('a')['href']).text, "html5lib")
        print('作者: ' + r.find(class_='article-meta-value').text)
    except:
        continue
```

### ② 印出第一頁所有文章的「作者」「標題」「時間」

```python
import pandas as pd

df = pd.DataFrame(columns=['poster', 'catagoly','topic','date'])
i=0

for d in soup.find_all(class_="title"):
    try:
        r = BeautifulSoup(requests.get('https://www.ptt.cc'+d.find('a')['href']).text, "html5lib")
        data = r.find_all(class_='article-meta-value')
        df.loc[i] = [data[0].text, data[1].text, data[2].text, data[3].text]
        i+=1
    except:
        continue
    
df
```

### ③ 試著爬爬看其他版的文章
